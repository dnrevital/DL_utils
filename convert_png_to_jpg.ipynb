## Imports

```python
import os
import cv2
```

## Settings

```python
input_dir = '/home/drevital/pallet_detection/feb23_task_test_1'
```

## Convert png to jpg

```python
png_files = [f for f in os.listdir(input_dir) if f.endswith('.png')]

for png_file in png_files:
    img = cv2.imread(os.path.join(input_dir, png_file))
    output_file = os.path.splitext(png_file)[0] + '.jpg'
    cv2.imwrite(os.path.join(input_dir, output_file), img, [int(cv2.IMWRITE_JPEG_QUALITY), 90])
```
