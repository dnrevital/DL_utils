## Imports

```python
import os
import shutil
```

## Settings

```python
eval_portion = 0.1 # Percentage of eval dataset from whole dataset, /100, range 0.0-1.0
eval_nth_select = int(1.0/eval_portion) # Each nth file from original list is moved to eval/
dataset_dir = '/home/drevital/obstacles_classification_datasets/rgb_6_K'
train_dir = os.path.join(dataset_dir, 'train')
eval_dir = os.path.join(dataset_dir, 'eval')
train_obs_dir = os.path.join(train_dir, 'obstacle')
train_no_obs_dir = os.path.join(train_dir, 'no_obstacle')
#train_sdv_dir = os.path.join(train_dir, 'SDV')
eval_obs_dir = os.path.join(eval_dir, 'obstacle')
eval_no_obs_dir = os.path.join(eval_dir, 'no_obstacle')
#eval_sdv_dir = os.path.join(eval_dir, 'SDV')
```

## Move every nth obstacle file to eval/obstacle

```python
orig_obs_files = os.listdir(train_obs_dir)

for i, f in enumerate(orig_obs_files):
    if i % eval_nth_select == 0:
        src = os.path.join(train_obs_dir, f)
        shutil.move(src, eval_obs_dir)
```

## Move every nth no_obstacle file to eval/no_obstacle

```python
orig_no_obs_files = os.listdir(train_no_obs_dir)

for i, f in enumerate(orig_no_obs_files):
    if i % eval_nth_select == 0:
        src = os.path.join(train_no_obs_dir, f)
        shutil.move(src, eval_no_obs_dir)
```
