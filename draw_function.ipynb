## Imports

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Settings

```python
x_range = [-1, 1]
k = 3
sigma = 2
```

# Curve function definition

```python
def f(x):    
    return np.exp(-(x-((k+1)**2)) / 2*(sigma**2))
```

# Main

```python
x = np.linspace(x_range[0], x_range[1], 100)
y = f(x)

plt.plot(x, y)
plt.show()
```
