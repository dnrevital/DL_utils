## Imports

```python
import os
import io
import boto3
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

## Settings

```python
imdir = '../obstacles_classification_datasets/test_3/train'
imname = '43_obstacles_classification_10_2022-11-09T13-34-51_1_0.03_shufersal_obs_3.0000_.jpg'
impath = os.path.join(imdir, imname)
```

## Utils

```python
def display_image(im, title):
    plt.imshow(im)
    plt.title(title)
    plt.show()    
```

## Main

```python
im = cv2.imread(os.path.join(imdir, imname))
im = cv2.cvtColor(im, cv2.COLOR_BGR2RGB) 
im = np.array(im, dtype='float32')
im /= 255.

# Divide im to original components (ref, current, mask)
w = im.shape[1]
ref = im[:, :w//3]
current = im[:, w//3:(w*2)//3] 
mask = im[:, (w*2)//3:] 

# Horizontal Flip
ref = cv2.flip(ref, 0)
current = cv2.flip(current, 0)

# Set mask to zero
h = mask.shape[0]
w = mask.shape[1]
mask = np.zeros((h, w, 3), dtype='float32')

# Build flipped image
flipped = cv2.hconcat([ref, current, mask])

# Display Images
display_image(im, 'im')
display_image(flipped, 'flipped')
```
