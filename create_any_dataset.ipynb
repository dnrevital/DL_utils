## Imports

```python
import os
import shutil
import numpy as np
```

## Definitions

```python
src_ds = '/home/drevital/obstacles_classification_datasets/suzuyo/eval'
dst_ds = '/home/drevital/obstacles_classification_datasets/new_factory/eval'
src_no_obs_ds = os.path.join(src_ds, 'no_obstacle')
#src_obs_ds = os.path.join(src_ds, 'obstacle')
dst_no_obs_ds = os.path.join(dst_ds, 'no_obstacle')
#dst_obs_ds = os.path.join(dst_ds, 'obstacle')
max_dst_each_class_images = 205
```

## List each class files and calculate no. of images to be prepared

```python
src_no_obs_imnames = os.listdir(src_no_obs_ds)
src_num_no_obs = len(src_no_obs_imnames)
dst_num_no_obs = min(src_num_no_obs, max_dst_each_class_images)
#src_obs_imnames = os.listdir(src_obs_ds)
#src_num_obs = len(src_obs_imnames)
#dst_num_obs = min(src_num_obs, max_dst_each_class_images)
#dst_num_images = min(dst_num_no_obs, dst_num_obs)
dst_num_images = dst_num_no_obs
```

## Choose randomally images to be prepared

```python
dst_no_obs_imnames = np.array(src_no_obs_imnames)
np.random.shuffle(dst_no_obs_imnames)
dst_no_obs_imnames = dst_no_obs_imnames[:dst_num_images]
dst_no_obs_imnames = list(dst_no_obs_imnames)
#dst_obs_imnames = np.array(src_obs_imnames)
#np.random.shuffle(dst_obs_imnames)
#dst_obs_imnames = dst_obs_imnames[:dst_num_images]
#dst_obs_imnames = list(dst_obs_imnames)
```
