```python
import tensorflow as tf
import numpy as np
```

```python
def custom_loss_1(y_true, y_pred):
    return tf.keras.metrics.categorical_crossentropy(y_true, y_pred)
```

```python
def custom_loss_2(y_true, y_pred):
    yt_1_2 = y_true[:, 1:]
    yp_1_2 = y_pred[:, 1:]
    
    return tf.keras.metrics.categorical_crossentropy(y_true, y_pred)\
         + tf.keras.metrics.categorical_crossentropy(yt_1_2, yp_1_2)
```

```python
y_true = np.array([[1, 0, 0], [0, 1, 0], [0, 0, 1], [0, 0, 1]])
y_pred = np.array([[0.9, 0.1, 0.0], [0.1, 0.4, 0.5], [0.1, 0.5, 0.4], [0.1, 0.5, 0.4]])
```

```python
custom_loss_1(y_true, y_pred)
```

```python
custom_loss_2(y_true, y_pred)
```
