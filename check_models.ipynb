```python
import tensorflow as tf
import tensorflow_hub as hub
```

```python
rcnn_1024 = 'https://hub.tensorflow.google.cn/tensorflow/mask_rcnn/inception_resnet_v2_1024x1024/1'
```

```python
rcnn_layer = hub.KerasLayer(rcnn_1024, trainable=False)
```

```python
model = tf.keras.Sequential()
```

```python
model
```

```python
model.layers[0].trainable_weigts
```

```python
model.add(rcnn_layer)
```

```python
num_classes = 10
model = tf.keras.Sequential([rcnn_layer, tf.keras.layers.Dense(num_classes)])
```
