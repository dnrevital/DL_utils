## Imports

```python
import os
import numpy as np
import cv2
```

## Settings

```python
imdir = '/home/drevital/obstacles_classification_datasets/new_factory/obstacle_pairs'
imname = 'y3_20_831_.98_2021-07-27--17-44-32_.jpg'
outdir = 'home/drevital/obstacles_classification_model'
img_height = 200
img_width = 200
```

## Main

```python
imrgb = cv2.imread(os.path.join(imdir, imname))
im = cv2.cvtColor(imrgb, cv2.COLOR_BGR2GRAY)  
w = im.shape[1]
im1 = im[:,:w//2]
im2 = im[:,w//2:]    
outpath1 = os.path.join(outdir, 'im1.jpg')
outpath2 = os.path.join(outdir, 'im2.jpg')
cv2.imwrite(outpath1, im1)
cv2.imwrite(outpath2, im2)
```
