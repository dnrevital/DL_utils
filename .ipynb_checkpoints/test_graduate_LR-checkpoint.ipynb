```python
import math
```

```python
def step_decay(epoch, lr):
   drop = 0.5
   epochs_drop = 2.0
   lrate = lr * math.pow(drop,  math.floor((epoch-1)/epochs_drop))
    
   return lrate
```

```python
lr = 0.001

for epoch in range(1, 11):
    print(f'epoch: {epoch}, learning rate: {step_decay(epoch, lr):.8f}')
```
