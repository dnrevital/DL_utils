## Imports

```python
import cv2
import os
import numpy as np
from matplotlib import pyplot as plt
```

# Settings

```python
folder = '/home/drevital/obstacles_classification_datasets/obstacle_classification_RGB_data/new_factory_rgb_data/all_data/classified_no_obstacle'
imname = '1_51_.74_2021-08-15--18-13-18_.jpg'
```

## Load Image

```python
im = cv2.imread(os.path.join(folder,imname))
```

## Split Images

```python
w = im.shape[1]
im1 = im[:,:w//2]
im2 = im[:,w//2:]
```

## Display Splitted Images

```python
plt.imshow(im1)
plt.title('left half = source')
plt.show()

plt.imshow(im2)
plt.title('right half = current')
plt.show()
```

## Convert each separated image to Green

```python
gs_im1 = cv2.cvtColor(im1, cv2.COLOR_BGR2GRAY) 
gs_im2 = cv2.cvtColor(im2, cv2.COLOR_BGR2GRAY)
g_im1 = im1[:, :, 1]
g_im2 = im2[:, :, 1]
```

## Display side-by-side with gray-scale

```python
fig, ax = plt.subplots(2, 2, figsize=(16,8))

ax[0][0].set_title('ref, gray')
ax[0][1].set_title('ref, green')
ax[1][0].set_title('cur, gray')
ax[1][1].set_title('cur, green')

ax[0][0].imshow(gs_im1, cmap='gray', vmin=0, vmax=255)
ax[0][1].imshow(g_im1, cmap='gray', vmin=0, vmax=255)
ax[1][0].imshow(gs_im2, cmap='gray', vmin=0, vmax=255)
ax[1][1].imshow(g_im2, cmap='gray', vmin=0, vmax=255)
```

## Merge the splitted images into one 2-channel image

```python
gg_im = cv2.merge([g_im1, g_im2])
```
