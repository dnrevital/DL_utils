```python
import tensorflow as tf
import tensorflow_hub as hub
```

```python
rcnn_1024 = 'https://hub.tensorflow.google.cn/tensorflow/mask_rcnn/inception_resnet_v2_1024x1024/1'
```

```python
rcnn_layer = hub.KerasLayer(rcnn_1024, input_shape=(1024, 1024, 3), trainable=False)
```

```python
rcnn_seq = tf.keras.Sequential([rcnn_layer])
```

```python
inputs = tf.keras.Input((1024, 1024, 3))
outputs = tf.keras.layers.Dense(5, activation=tf.nn.softmax)(rcnn_seq)
```
