## Imports

```python
import os
import numpy as np
import cv2
```

## Settings

```python
B = 0
G = 127
R = 255
height = 200
width = 400
classes = ['no_obstacle', 'obstacle']
im_per_class = 100
root_dir = '/home/drevital/obstacles_classification_datasets/synt'
```

## Build Synthetic Image

```python
synt = [[[B, G, R] for j in range(width)] for i in range(height)]
synt = np.array(synt)
```

```python
synt.shape
```

```python
synt[:10]
```

## Store synthetic images

```python
for c in classes:
    path = os.path.join(root_dir, c)
    for i in range(im_per_class):
        imname = f'synt_{i}.jpg'
        impath = os.path.join(path, imname)
        cv2.imwrite(impath, synt)
```
