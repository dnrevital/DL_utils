```python
from collections import defaultdict
```

```python
inds = defaultdict(list)
```

```python
nums = [11, 20, 31, 4, 42, 100, 201, 6, 77, 66, 55]
target = 10
```

```python
for i, n in enumerate(nums):
    inds[n] = [i]
```

```python
inds
```

```python
subs = [target-n for n in nums]
```

```python
subs
```

```python
for j, s in enumerate(subs):
    inds[s] += [j]
```

```python
for val in inds.values():
    if len(val) == 2:
        print(f'The indices are {val[0]} and {val[1]}')
        break
```
