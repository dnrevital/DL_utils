```python
class LRDecayCallback():
    def __init__(self):
        self.min_lr = 5e-6
        self.max_lr = 5e-5
        self.min_loss = 0.85
        self.max_loss = 37.5
        self.loss_range = self.max_loss - self.min_loss
        self.lr_range = self.max_lr - self.min_lr
        
    def on_epoch_end(self, val_loss):
        f = (val_loss - self.min_loss) / self.loss_range
        lr = self.min_lr + f * self.lr_range
        print(f'LR: {lr:.8f}')
```

```python
val_losses = [37.5, 30.0, 21.0, 9.0, 2.0, 1.5, 1.0, 0.85]
dec = LRDecayCallback()
for val_loss in val_losses:
    dec.on_epoch_end(val_loss)
```
