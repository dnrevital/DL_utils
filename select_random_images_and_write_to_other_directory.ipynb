## Imports

```python
import os
import shutil
import numpy as np
from numpy.random import default_rng
from pathlib import Path
```

## Settings

```python
src_dir = '/home/drevital/obstacles_classification_datasets/3_class/pairs/obstacle_1'
dst_dir = '/home/drevital/obstacles_classification_datasets/3_class/pairs/augment_obstacle'
num_copied = 20
```

## Main

```python
Path(dst_dir).mkdir(parents=True, exist_ok=True)
fnames = os.listdir(src_dir)
num_files = len(fnames)
rng = default_rng()
rands = rng.choice(num_files, size=num_copied, replace=False)
```

```python
for i in rands:
    src = os.path.join(src_dir, fnames[i])
    dst_imname = '.'.join(fnames[i].split('.')[:-1]) + '_' + '.jpg'
    dst = os.path.join(dst_dir, dst_imname)
    shutil.copy(src, dst)
```
