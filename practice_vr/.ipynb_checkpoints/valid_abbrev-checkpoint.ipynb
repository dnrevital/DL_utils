```python
def is_num(c):
    return ord(c) in range(ord('0'), ord('9')+1)

def valid_abbr(word, abbr):
    word_size = len(word)
    abbr_size = len(abbr)
    i = 0
    j = 0
    
    while i < word_size and j < abbr_size:
        print(f'word[{i}]: {word[i]}')
        print(f'abbr[{j}]: {abbr[j]}')
        if not is_num(abbr[j]):
            if word[i] != abbr[j]:
                return False
            i += 1
        else:
            n = int(abbr[j])
            if n == 0:
                return False
            if i + n >= word_size:
                return False
            if j < abbr_size-1 and is_num(abbr[j+1]):
                return False
            
            i += n
        j += 1
            
    return True
```
