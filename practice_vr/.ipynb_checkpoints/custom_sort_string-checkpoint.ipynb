```python
from collections import Counter

def custom(order, s):
    counts = Counter(s)
    ordered = ''
    
    for c in order:
        ordered += counts[c]*c
        
    for c in s:
        if c not in order:
            ordered += c
            
    return ordered
```

```python
def custom_with_sort(order, s):
    inds = {c: i for i, c in enumerate(order)}
    for c in s:
        if c not in inds.keys():
            inds[c] = len(s)
    vals = {i: c for c, i in inds.items()}
    nums = [inds[c] for c in s]    
    nums.sort()
    
    return ''.join([vals[num] for num in nums])            
```
