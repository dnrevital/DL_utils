```python
def diameter(root):
    return max(recurse(root))

def recurse(root):
    if not root.left and not root.right:
        return [0, 0]
    
    up_to_left = up_to_right = max_left = max_right = 0
    
    if root.left:
        up_to_left, max_left = recurse(root.left)
    if root.right:
        up_to_right, max_right = recurse(root.right)
        
    return [max(up_to_left, up_to_right) + 1,
            max(max_left, max_right, up_to_left + up_to_right + 2)]        
```

```python
class Node:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None
```

```python
root = Node(1)
root.left = Node(2)
root.left.left = Node(4)
root.left.right = Node(5)
root.right = Node(3)
```
