```python
from collections import defaultdict

def shifted_group(strings):
    shifted = defaultdict(list)
    
    for s in strings:
        key = '0'
        for i, _ in enumerate(s[1:]):
            key += str((ord(s[i+1])-ord(s[i]))%26)
        shifted[key].append(s)
        
    return list(shifted.values())
```
