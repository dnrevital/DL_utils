```python
def is_bst(tree):
    left_cond, right_cond = True, True
    
    if tree.left:
        left_cond = is_bst(tree.left) and tree.left.val <= tree.val
    if tree.right:
        right_cond = is_bst(tree.right) and tree.right.val >= tree.val
        
    return left_cond and right_cond
```
