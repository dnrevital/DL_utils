```python
def flip_equiv(t1, t2):
    if not t1 and not t2:
        return True
    if not t1 or not t2:
        return False
    
    return (t1.val == t2.val) and\
        ((flip_equiv(t1.left, t2.left) and flip_equiv(t1.right, t2.right)) or\
        (flip_equiv(t1.left, t2.right) and flip_equiv(t1.right, t2.left)))
```

```python
class Node:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None
```

```python
t1 = Node(1)
t1.left = Node(2)
t1.left.left = Node(4)
t1.left.right = Node(5)
t1.left.right.left = Node(7)
t1.left.right.right = Node(8)
t1.right = Node(3)
t1.right.left = Node(6)

t2 = Node(1)
t2.left = Node(3)
t2.left.right = Node(6)
t2.right = Node(2)
t2.right.left = Node(4)
t2.right.right = Node(5)
t2.right.right.left = Node(8)
t2.right.right.right = Node(7)                         
```
