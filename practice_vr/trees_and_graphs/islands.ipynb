```python
def islands(grid):
    h = len(grid)
    w = len(grid[0])
    count = 0
    visited = [[False for j in range(w)] for i in range(h)]
    
    for i in range(h):
        for j in range(w):
            if grid[i][j] == '1' and not visited[i][j]:
                count += 1
                recurse_islands(i, j, grid, visited)
                
    return count
```

```python
def recurse_islands(i, j, grid, visited):
    h = len(grid)
    w = len(grid[0])
    
    visited[i][j] = True    
    nbrs = []
    if i > 0:
        nbrs.append([i-1, j])
    if j > 0:
        nbrs.append([i, j-1])
    if i < h-1:
        nbrs.append([i+1, j])
    if j < w-1:
        nbrs.append([i, j+1])
        
    for nbr in nbrs:
        if grid[nbr[0]][nbr[1]] == '1' and not visited[nbr[0]][nbr[1]]:
            recurse_islands(nbr[0], nbr[1], grid, visited)           
```
