```python
def zigzag(tree):
    nodes = [[tree.val]]
    recurse_zigzag(tree, nodes, ['right', 'left'])
    
    return nodes

def recurse_zigzag(tree, nodes, order):
    if order[0] == 'left':
        next_nodes = [tree.left, tree.right]
    else:
        next_nodes = [tree.right, tree.left]
        
    next_nodes_list = []
        
    if next_nodes[0]:
        next_nodes_list.append(next_nodes[0].val)
    if next_nodes[1]:
        next_nodes_list.append(next_nodes[1].val)
    if next_nodes_list:
        nodes.append(next_nodes_list)
        
    order[0], order[1] = order[1], order[0]
    if next_nodes[0]:
        recurse_zigzag(next_nodes[0], nodes, order)
    if next_nodes[1]:
        recurse_zigzag(next_nodes[1], nodes, order)       
```

```python
class Node:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None
```

```python
tree = Node(3)
tree.left = Node(9)
tree.right = Node(20)
tree.right.left = Node(15)
tree.right.right = Node(7)
```
