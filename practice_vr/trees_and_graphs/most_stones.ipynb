```python
def most_stones_1(stones):
    num_stones = len(stones)
    visited = [False for stone in stones]
    islands = 0
    
    for i, stone in enumerate(stones):
        if not visited[i]:
            islands += 1
            dfs(i, stones, visited)
            
    return num_stones - islands
            
def dfs(i, stones, visited):
    visited[i] = True
    cur_stone = stones[i]
    
    for j, stone in enumerate(stones):
        if j == i:
            continue
        if not visited[j] and (stone[0] == cur_stone[0] or stone[1] == cur_stone[1]):
            dfs(j, stones, visited)
```
