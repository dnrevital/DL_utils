```python
nums = [0,0,1,1,1,2,2,3,3,4]
```

```python
def swap(nums, i, j):
    tmp = nums[i]
    nums[i] = nums[j]
    nums[j] = tmp

def rem_duplicates(nums):
    cur_ind = 0
    last_num = nums[0]
    size = len(nums)
    
    for i in range(1, size):
        if nums[i] > last_num:
            last_num = nums[i]
            cur_ind += 1
            if cur_ind != i:
                swap(nums, cur_ind, i)
            
    return cur_ind + 1
```

```python
rem_duplicates(nums)
```
