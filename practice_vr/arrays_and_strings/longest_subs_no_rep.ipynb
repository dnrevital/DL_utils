```python
def max_no_rep(s):
    if not s:
        return 0
    
    max_len = 1
    start = 0
    cur = 1
    
    for c in s[1:]:
        if c in s[start:cur]:
            start += 1
        l = cur + 1 - start
        if l > max_len:
            max_len = l
        cur += 1
            
    return max_len        
```

```python
def longest_no_rep(s):
    size = len(s)
    if size == 1:
        return 1
    start = 0
    longest = 0
    
    for i, c in enumerate(s[1:]):
        if c in s[start:i+1]:
            start += 1
        else:
            longest = max(longest, i-start+2)
        
    return longest
```
