```python
def sent(emails):
    sent = []
    
    for email in emails:
        local, domain = email.split('@')
        if domain[-4:] != '.com':
            continue
        domain = domain[:-4]
        if '.' in domain or '+' in domain:
            continue
        local = local.split('+')[0]
        local = ''.join(local.split('.'))
        sent.append(local + '@' + domain + '.com')
        
    return len(sent)
```
