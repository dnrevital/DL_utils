```python
def disappear(nums):
    size = len(nums)
    appear = {i: False for i in range(1, size+1)}
    
    for n in nums:
        appear[n] = True
        
    return [i for i in range(1, size+1) if not appear[i]]
```

```python
nums = [4,3,2,7,8,2,3,1]
disappear(nums)
```
