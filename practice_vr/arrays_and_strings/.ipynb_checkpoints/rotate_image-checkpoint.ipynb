```python
def swap4(nums):
    print(f'nums: {nums}')
    print(f'returned: {nums[3], nums[0], nums[1], nums[2]}')
    return [nums[3], nums[0], nums[1], nums[2]]
```

```python
def rotate(matrix):
    size = len(matrix)
    
    for i in range(size//2):
        top = i
        bottom = size-i-1
        left = i
        right = size-i-1
        
        for j in range(left, right):
            to_swap = [matrix[top][left+j],
                       matrix[top+j][right],
                       matrix[bottom][right-j],
                       matrix[bottom-j][left]]
            swapped = swap4(to_swap)
            matrix[top][left+j] = swapped[0]
            matrix[top+j][right] = swapped[1]
            matrix[bottom][right-j] = swapped[2]
            matrix[bottom-j][left] = swapped[3]
```

```python
matrix = [[5,1,9,11],[2,4,8,10],[13,3,6,7],[15,14,12,16]]
```

```python
rotate(matrix)
```
