```python
letters = {'2': ['a', 'b', 'c'],
           '3': ['d', 'e', 'f'],
           '4': ['g', 'h', 'i'],
           '5': ['j', 'k', 'l'],
           '6': ['m', 'n', 'o'],
           '7': ['p', 'q', 'r', 's'],
           '8': ['t', 'u', 'v'],
           '9': ['w', 'x', 'y', 'z']}
```

```python
def combs(nums):
    if not nums:
        return None
    
    if len(nums) == 1:
        return [letter for letter in letters[nums[0]]]
                
    return [letter + comb for comb in combs(nums[1:]) for letter in letters[nums[0]]]
```

```python
nums = '23'
combs(nums)
```
