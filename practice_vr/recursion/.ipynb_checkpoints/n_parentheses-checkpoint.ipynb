```python
def par(n):
    if n == 0:
        return ['']
        
    if n == 1:
        return ['()']

    options = []
    for i in range(1, n+1):
        left = n-i
        for p1 in par(i-1):
            s1 = '(' + p1 + ')'
            for s2 in par(left):
                options.append(s1 + s2)
        
    return options
```
