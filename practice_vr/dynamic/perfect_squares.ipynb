```python
import math
import sys

def perfect_squares(n):
    root = math.sqrt(n)
    froot = math.floor(root)
    ps = [i**2 for i in range(froot, 0, -1)]

    return rec_ps(n, ps)

def rec_ps(n, ps):
    sum1 = sum2 = sys.maxsize

    if len(ps) == 1:
        if n == ps[0]:
            return 1
        return -1

    div = n//ps[0]
    rem = n % ps[0]

    if not rem:
        return div

    comb1 = div + rec_ps(rem, ps[1:])
    comb2 = rec_ps(n, ps[1:])

    if comb1 != -1:
        sum1 = comb1
    if comb2 != -1:
        sum2 = comb2

    return min(sum1, sum2)    
```
