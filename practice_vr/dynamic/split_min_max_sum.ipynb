```python
import sys

def part_min_max(nums, m):
    size = len(nums)
    
    if size == 1:
        return nums[0]    
    if m == 1:
        return(sum(nums))    
    if m == size:
        return max(nums)
    
    max_grp_size = size-m+1
    min_max_sum = sys.maxsize
    grp_sum = 0
    
    for grp_size in range(1, max_grp_size+1):
        grp = nums[:grp_size]
        grp_sum += nums[grp_size-1]
        max_sum = max(grp_sum, part_min_max(nums[grp_size:], m-1))
        if max_sum < min_max_sum:
            min_max_sum = max_sum
            
    return min_max_sum
```
