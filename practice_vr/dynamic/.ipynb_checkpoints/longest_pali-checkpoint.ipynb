```python
def longest_pali(s):
    max_len = 1
    size = len(s)
            
    for i in range(size-1):
        if s[i] == s[i+1]:

            cur_len = 0
            cur_i = i
            cur_j = i+1

            while cur_i >= 0 and cur_j < size and s[cur_i] == s[cur_j]:
                cur_len += 2
                cur_i -= 1
                cur_j += 1

            max_len = max(cur_len, max_len)
        
    for i in range(size-2):

        if s[i] == s[i+2]:
            cur_len = 1
            cur_i = i
            cur_j = i+2

            while cur_i >= 0 and cur_j < size and s[cur_i] == s[cur_j]:
                cur_len += 2
                cur_i -= 1
                cur_j += 1

            max_len = max(cur_len, max_len)
                
    return max_len
```
