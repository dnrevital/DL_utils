```python
def min_coins(coins, amount):
    coins = sorted(coins, reverse=True)
    
    return recurse_min_coins(coins, amount)

def recurse_min_coins(coins, amount):
    size = len(coins)
    
    if size == 1:
        if amount == coins[0]:
            return 1
        return -1
    
    div = amount // coins[0]
    remainder = amount % coins[0]
    rest = recurse_min_coins(coins[1:], remainder)
    if rest == -1:
        return recurse_min_coins(coins[1:], remainder)
    return div + rest
    
```
