```python
def longest_sub(nums):
    if not nums:
        return 0
    
    size = len(nums)
    
    length = [[0 for j in range(size)] for i in range(size)]
    
    for i in range(size):
        length[i][i] = 1
        
    longest = 1
        
    for gap in range(1, size):
        for i in range(size-gap):
            j = i + gap
            length[i][j] = length[i][j-1]
            if nums[j] > nums[j-1]:
                length[i][j] = length[i][j-1] + 1
            if length[i][j] > longest:
                longest = length[i][j]
        
    return longest
```
