```python
import sys

def max_sub(nums):
    size = len(nums)
    sums = [[0 for i in range(size)] for j in range(size)]
    max_sum = -sys.maxsize-1
    
    for i in range(size):
        sums[i][i] = nums[i]
        
    for i in range(size):
        for j in range(1, size-i-1):
            cur_sum = sums[i][i+j-1] + nums[i+j]
            sums[i][i+j] = cur_sum
            if cur_sum > max_sum:
                max_sum = cur_sum
                
    return max_sum
```

```python
nums = [-2,1,-3,4,-1,2,1,-5,4]
max_sub(nums)
```
