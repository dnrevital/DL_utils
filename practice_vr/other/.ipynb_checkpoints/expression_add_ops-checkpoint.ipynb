```python
def add_ops(nums, target):
    
    if not nums:
        return []
    
    if len(nums) == 1:
        if nums[0] == target:
            return str(nums[0])
        return []

    elements = [nums[0]]    
    element = nums[0]
    
    for n in nums[1:]:
        element *= n
        elements.append(element)
    
    ret = []
    prefix = str(nums[0])
    start = 1
    
    for element in elements[:-1]:
        plus = add_ops(nums[start:], target-element)
        minus_nums = [-nums[start]] + nums[start+1:]
        minus = add_ops(minus_nums, target-element)
        
        orig_s0 = ''

        for s in plus:
            ret.append(prefix + '+' + s)
        for s in minus:
            if len(s) < 2:
                break
            if s[0] == '-':
                orig_s0 = -(int(s[0:2]))
            orig_s = str(orig_s0)
            if len(s) > 2:
                orig_s += s[2:]
            ret.append(prefix + '-' + orig_s)
        
        prefix = prefix + '*' + str(nums[start])
        start += 1
        
    if elements[-1] == target:
        ret.append(prefix)
               
    if not ret:
        return []
    
    return ret
```

```python
nums = [1, 2, 3, 4, 5]
target = 6
add_ops(nums, target)
```
