```python
import sys

def int_val(s):
    if s == '':
        return sys.maxsize
    return int(s)

def remove_k(num, k):
    size = len(num)
    
    if k == size:
        return ''
    
    if not k:
        return num
    
    smallest = min(int_val(num[0] + remove_k(num[1:], k)), int_val(remove_k(num[1:], k-1)))
    
    return str(smallest)
```
