```python
def calc(s):
    
    # Tokenize
    size = len(s)
    i = 0
    tokens = []
    num = ''
    
    while i < size:
        while i < size and ord(s[i]) in range(ord('0'), ord('9')+1):
            num += s[i]
            i += 1
        if num:
            if tokens and tokens[-1]=='-' and (len(tokens)==1 or tokens[-2] in ['+', '-']):
                tokens[-1] = '-' + num
            else:
                tokens.append(num)
                num = ''
        if i < size and s[i] in ['+', '-', '(', ')']:
            tokens.append(s[i])
            
        i += 1
                    
    res = 0
    stack = []
    sign = 1
    op_sign = 1

    for i, token in enumerate(tokens):
        if stack:
            sign = stack[0]
        else:
            sign = 1

        if ord(token[0]) in range(ord('0'), ord('9')+1):
            res += op_sign * sign * int(token)
        elif token == '+':
            op_sign = 1
        elif token == '-':
            op_sign = -1
        elif token == '(':
            last_sign = 1
            if i > 0:
                last_sign = 1 if tokens[i-1] == '+' else -1
            stack_sign = 1
            if stack:
                stack_sign = stack[0]
            stack.insert(0, last_sign * stack_sign)
        else:
            stack.pop(0)
                                
    return res       
```
