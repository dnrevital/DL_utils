```python
def cycle(head):
    visited = []
    cur = head
    
    while cur:
        if cur.val in visited:
            return True
        visited.append(cur.val)
        cur = cur.next
        
    return False
```

```python
class Node:
    def __init__(self, val):
        self.val = val
        self.next = None
```

```python
head = Node(3)
head.next = Node(2)
head.next.next = Node(0)
head.next.next.next = Node(4)
head.next.next.next.next = head.next
```
