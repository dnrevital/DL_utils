```python
def remove(linked, n):
    count = 0
    cur = linked
    
    while cur:
        count += 1
        cur = cur.next
        
    target = count - n
    
    if target == 0:
        return linked.next
    
    cur = linked
    count = 0
    
    print(f'target: {target}')
    
    while count < target-1:
        count += 1
        cur = cur.next
        
    cur.next = cur.next.next
    
    return linked
```

```python
class Node:
    def __init__(self, val):
        self.val = val
        self.next = None
```

```python
linked = Node(1)
linked.next = Node(2)
linked.next.next = Node(3)
linked.next.next.next = Node(4)
linked.next.next.next.next = Node(5)
```
