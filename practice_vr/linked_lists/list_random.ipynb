```python
class Node:
    def __init__(self, val):
        self.val = val
        self.next = None
        self.random = None
```

```python
def clone(head):
    pointing = {}
    pointed = {}
    
    # Fill dictionary of old clones
    
    cur = head
    ind = 0
    pointing = {}
    
    while cur:
        if cur.random:
            pointing[ind] = cur.random
        cur = cur.next
        ind += 1
        
    cur_old = head
    new_head = None
    ind = 0
    prev_new = None
    
    while cur_old:
        cur_new = Node(cur_old.val)
        if not new_head:
            new_head = cur_new
        if ind in pointing.keys():
            pointed[ind] = cur_new
        if prev_new:
            prev_new.next = cur_new
        prev_new = cur_new
        cur_old = cur_old.next
        ind += 1
        
    cur = head
    cur_new = new_head
    
    while cur:
        if cur in pointing.values():
            pointing_list = list(pointing.values())
            pointing_keys = list(pointing.keys())
            pointing_key_inds = [i for i, node in enumerate(pointing_list) if node==cur]
            for pointing_key_ind in pointing_key_inds:
                pointed_ind = pointing_keys[pointing_key_ind]
                pointed[pointed_ind].random = cur_new
        cur = cur.next
        cur_new = cur_new.next
        
    return new_head
```
