```python
class Node:
    def __init__(self, val):
        self.val = val
        self.next = None
        self.previous = None
        
class Linked:
    def __init__(self, head):
        self.head = head
        
    def print_nodes(self):
        cur = self.head
        print('\n')
        
        while cur:
            print(f'{cur.val}==>', end=' ')
            cur = cur.next
        
    def reverse(self):
        cur = self.head

        while cur.next:
            cur.next.previous = cur
            cur = cur.next

        self.head = cur
        
        while cur:
            print(f'cur: {cur.val}')
            cur.next = cur.previous
            cur = cur.next
```

```python
linked = Linked(Node(1))
linked.head.next = Node(2)
linked.head.next.next = Node(3)
linked.head.next.next.next = Node(4)
linked.head.next.next.next.next = Node(5)

linked.print_nodes()
linked.reverse()
linked.print_nodes()
```
