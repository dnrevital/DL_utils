```python
def merge(list1, list2):
    if list1.head.val <= list2.head.val:
        a = list1
        b = list2
    else:
        a = list2
        b = list1
        
    cur = a.head
    
    while cur.next and b.head:
        if b.head.val >= cur.next.val:
            cur = cur.next
        else:
            tmp = b.head
            b.head = b.head.next
            tmp.next = cur.next
            cur.next = tmp
            
    while b.head:
        tmp = b.head
        b.head = b.head.next
        tmp.next = cur.next
        cur.next = tmp
        cur = cur.next
        
    return a.head
```

```python
class Node:
    def __init__(self, val):
        self.val = val
        self.next = None
        
class Linked:
    def __init__(self, head):
        self.head = head
        
    def dump(self):
        cur = self.head
        
        while cur:
            print(f'{cur.val}==>', end=' ')
            cur = cur.next
```

```python
list1 = Linked(Node(1))
list1.head.next = Node(2)
list1.head.next.next = Node(4)
print('list1:')
list1.dump()

list2 = Linked(Node(1))
list2.head.next = Node(3)
list2.head.next.next = Node(4)
print('\nlist2:')
list2.dump()

merged = merge(list1, list2)
merged_list = Linked(merged)
print('\nmerged:')
merged_list.dump()
```
