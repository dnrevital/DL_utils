```python
def first_last_pos(nums, target):
    size = len(nums)
    left = 0
    right = size-1
    cur = size // 2
    first = None
    
    if size == 1:
        return [0, 0]
    
    while left < right:
        if nums[cur] == target and (cur==0 or nums[cur] > nums[cur-1]):
            first = cur
            break
        elif nums[cur] >= target:
            right = cur - 1
        elif cur < size-1:
            left = cur + 1
        cur = left + (right-left)//2
        
    if nums[cur] == target:
        first = cur
        
    if first is None:
        return [-1, -1]
    
    right = size-1
    size = right - first
    left = first
    cur = size // 2
    
    while left < right:
        if nums[cur] == target and (cur==size-1 or nums[cur] < nums[cur+1]):
            return [first, cur]
        if nums[cur] > target:
            right = cur - 1
        else:
            left = cur + 1                
        cur = left + (right-left)//2
        
    return [first, right]
```

```python
nums = [5,7,7,8,8,8,8,10,12,15]
target = 8
```

```python
first_last_pos(nums, target)
```
