```python
def peak(arr):
    if len(arr) == 3:
        return 1
    
    left = 0
    right = len(arr)-1
    size = right - left + 1
    
    while size > 3: 
        cur = left + (right - left)//2

        if arr[cur-1] < arr[cur] > arr[cur+1]:
            return cur
        if arr[cur-1] > arr[cur]:
            right = cur
        else:
            left = cur
            
        size = right - left + 1  
        if size == 3:
            return left + 1
```
