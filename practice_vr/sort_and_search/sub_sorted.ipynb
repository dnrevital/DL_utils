```python
def sub_sorted(arr):
    size = len(arr)
    
    # Divide into sorted groups
    groups = []
    group = [arr[0]]
    if size == 1:
        groups.append(group)
        
    for i in range(1, size):
        if arr[i] >= arr[i-1]:
            group.append(arr[i])
            if i == size-1:
                groups.append(group)
        else:
            groups.append(group)
            group = [arr[i]]
            if i == size-1:
                groups.append(group)

    return merge(groups)
```
