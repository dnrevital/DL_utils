```python
def non_overlap(intervals):
    max_end = max(interval[1] for interval in intervals)
    occupied = [False for i in range(max_end+1)]
    
    for interval in intervals:
        for i in range(interval[0], interval[1]+1):
            occupied[i] = True
            
    merged = []
    start = -1
    end = -1
    merged_size = len(occupied)
    if occupied[0]:
        start = 0
    
    for i in range(1, merged_size):
        if not occupied[i-1] and occupied[i]:
            start = i
        elif occupied[i-1] and not occupied[i]:
            end = i-1
            merged.append([start, end])
            
    if occupied[merged_size-1]:
        merged.append([start, merged_size-1])
        
    return merged
```
