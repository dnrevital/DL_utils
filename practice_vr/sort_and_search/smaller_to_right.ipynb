```python
def left_smaller(nums):
    size = len(nums)
    lsmaller = [0]
    count = 0
    
    for i in range(size-2, -1, -1):
        if nums[i] > nums[i+1]:
            count += 1
        lsmaller.insert(0, count)
        
    return lsmaller
```
