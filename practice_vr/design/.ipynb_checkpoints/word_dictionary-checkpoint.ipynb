```python
class Node:
    def __init__(self, char=''):
        self.subnodes = {'.': Node()}
        if char :
            self.char = char
```

```python
class WordDict:
    def __init__(self):
        self.root = Node()
        
    def addWord(self, word):
        cur = self.root
        prev = None
        
        for c in word:
            new = Node()
            if c not in cur.subnodes.keys():
                cur.subnodes[c] = new
            if prev:
                prev.subnodes['.'].subnodes[c] = new
            prev = cur
            cur = new
            
    def serach(self, word):
        cur = self.root
        
        for c in word:
            if c not in cur.subnodes.keys():
                return False
            cur = cur.subnodes[c]
            
        return True
```
