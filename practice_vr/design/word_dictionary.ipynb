```python
class WordDict:
    def __init__(self):
        self.root = {'.': {}}
        
    def addWord(self, word):
        cur = self.root
        prev = None
        
        for c in word:
            new = {}
            dot = {}

            if c not in cur.keys():
                cur[c] = new
            if '.' not in cur.keys():
                cur['.'] = dot
            if prev:
                prev['.'][c] = new
                prev['.']['.'] = dot
            prev = cur
            cur = new
            
    def search(self, word):
        cur = self.root
        
        for c in word:
            if c not in cur.keys():
                return False
            cur = cur[c]
            
        return True
```

```python
w = WordDict()
```

```python
w.addWord('bad')
```

```python
w.addWord('dad')
```

```python
w.addWord('mad')
```

```python
w.search('pad')
```

```python
w.search('bad')
```

```python
w.search('.ad')
```
