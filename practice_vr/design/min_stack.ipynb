```python
import heapq

class MinStack:
    def __init__(self):
        self.stack = []
        self.heap = []
        
    def push(self, val):
        self.stack.insert(0, val)
        heapq.heappush(self.heap, val)
        
    def pop(self):
        self.stack.pop(0)
        heapq.heappop(self.heap)
        
    def top(self):
        return self.stack[0]
        
    def get_min(self):
        return heapq.nsmallest(1, self.heap)[0]
```

```python
minStack = MinStack()
minStack.push(-2)
minStack.push(0)
minStack.push(-3)
minStack.get_min() # return -3
```

```python
minStack.pop()
minStack.top()    # return 0
```

```python
minStack.get_min() # return -2
```
