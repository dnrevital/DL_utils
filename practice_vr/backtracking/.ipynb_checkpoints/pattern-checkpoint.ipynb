```python
def is_matching(s, pattern):
    if len(s) == 1:
        return pattern in [s[0], '.', '*']
    
    if pattern[0] == '.':
        return len(pattern) > 1 and is_matching(s[1:], pattern[1:])
    
    if pattern[0] == '*':
        return len(pattern) == 1 or is_matching(s, pattern[1:]) or is_matching(s[1:], pattern[1:])
    
    return len(pattern) > 1 and s[0] == pattern[0] and is_matching(s[1:], pattern[1:])
```
