```python
my_widgets = []
my_widgets.append(widgets.Dropdown(options=['1', '2', '3']))
my_widgets.append(widgets.Dropdown(options=[('One', '1'), ('Two', '2'), ('Three', '3')]))
my_widgets.append(widgets.RadioButtons(options=['Obstacle', 'No Obstacle', 'Static', 'Removed']))
my_widgets.append(widgets.Select(options=['Obstacle', 'No Obstacle', 'Static', 'Removed']))
my_widgets.append(widgets.ToggleButtons(options=['Obstacle', 'No Obstacle', 'Static', 'Removed']))

display(*my_widgets)
```

```python
tog = widgets.ToggleButtons(options=['Obstacle', 'No Obstacle', 'Static', 'Removed'])
output = widgets.Output()
display(tog, output)

def annotate(change):
    with output:
        print(change['new'])
        
tog.observe(annotate, names='value')
```
