```python
import cv2
import io
import boto3
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
fig = plt.figure()

client = boto3.client("s3")
bucket='obstacles-classification'

key= 'test_dataset_3/obstacle/put1.jpg'
outfile = io.BytesIO()
client.download_fileobj(bucket, key, outfile)
outfile.seek(0)
im = plt.imread(outfile, format='jpg')
image = plt.imshow(im, cmap='gray', vmin=0, vmax=255)
fig.canvas.draw()

key= 'test_dataset_3/obstacle/put2.jpg'
outfile = io.BytesIO()
client.download_fileobj(bucket, key, outfile)
outfile.seek(0)
im = plt.imread(outfile, format='jpg')
image = plt.imshow(im, cmap='gray', vmin=0, vmax=255)
fig.canvas.draw()
```
