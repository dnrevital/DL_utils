## Imports

```python
import boto3
import io
import cv2
from matplotlib import pyplot as plt
```

```python
imrgb = cv2.imread('../SDV_Phase_Two/video_processor/new_factory_3/classified_obstacles/54_1591_2021-05-24--10-22-48.jpg')
im = cv2.cvtColor(imrgb, cv2.COLOR_BGR2GRAY)        
plt.imshow(im, cmap='gray', vmin=0, vmax=255)
plt.title('left half = source')
plt.show()
```

```python
bucket='obstacles-classification'
key= 'test_dataset_1/obstacle/test1.jpg'
client = boto3.client("s3")
```

```python
_,encoded = cv2.imencode('.jpg', im)
im_iobuf = io.BytesIO(encoded)
client.put_object(Body=im_iobuf, Key=key, Bucket=bucket)
```
