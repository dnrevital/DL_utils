## Imports

```python
import boto3
import io
import cv2
from matplotlib import pyplot as plt
```

```python
im = cv2.imread('/home/drevital/obstacles_classification_datasets/suzuyo/train/obstacle_pairs/1_98_2021-07-01--10-38-59.jpg')
plt.imshow(im, cmap='gray', vmin=0, vmax=255)
plt.title('left half = source')
plt.show()
```

```python
bucket='obstacles-classification'
key= 'test_dataset_1/obstacle/test3.jpg'
client = boto3.client("s3")
```

```python
_,encoded = cv2.imencode('.jpg', im)
im_iobuf = io.BytesIO(encoded)
client.put_object(Body=im_iobuf, Key=key, Bucket=bucket)
```
