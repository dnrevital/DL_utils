## Imports

```python
import os
import shutil
```

## Settings

```python
augmented_dir = '/home/drevital/obstacles_classification_datasets/gg_2_channels/obstacle_pairs'
dst_dir = '/home/drevital/obstacles_classification_datasets/rgb_6/annotated/obstacle'
aug_infixes = ['scale', 'rotate', 'noise', 'crop', 'shear', 'contrast', 'flip']
```

## Clean Augments

```python
fnames = os.listdir(augmented_dir)
imnames = ['.'.join(fname.split('.')[:-1]) for fname in fnames]
augmented = [any([aug in imname.split('_') for aug in aug_infixes]) for imname in imnames]

for i, fname in enumerate(fnames):
    if not augmented[i]:
        src = os.path.join(augmented_dir, fname)
        shutil.move(src, dst_dir)
```
