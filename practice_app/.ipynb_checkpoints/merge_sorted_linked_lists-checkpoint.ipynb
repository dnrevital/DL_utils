```python
class Linked:
    def __init__(self, n):
        self.n = n
        self.next = None
        
    def insert(self, n):
        self.next = Linked(n)
        
    def print(self):
        cur = self
        while cur:
            print(f'n: {cur.n} ==> ', end='')
            cur = cur.next            
        print()        
```

```python
l1 = Linked(1)
l1.insert(2)
l1.next.insert(4)

l2 = Linked(1)
l2.insert(3)
l2.next.insert(4)
```
