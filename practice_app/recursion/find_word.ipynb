```python
board = [["A","B","C","E"],
         ["S","F","C","S"],
         ["A","D","E","E"]]

word = "ABCCED"
```

```python
def find_word(board, word):
    rows = len(board)
    cols = len(board[0])
    visited = [[False for j in range(cols)] for i in range(rows)]
    
    for i in range(rows):
        for j in range(cols):
            if not visited[i][j]:
                if located(board, word, i, j, visited):
                    return True
                
    return False
```

```python
def located(board, word, i, j, visited):
    visited[i][j] = True
    rows = len(board)
    cols = len(board[0])
    
    nbrs = []
    
    if i > 0:
        nbrs.append([i-1, j])
    if i < rows-1:
        nbrs.append([i+1, j])
    if j > 0:
        nbrs.append([i, j-1])
    if j < cols-1:
        nbrs.append([i, j+1])
        
    for nbr in nbrs:
        rest = len(word)==1 or located(board, word[1:], nbr[0], nbr[1], visited)
        if board[i][j] == word[0] and rest:
            return True
        
    return False
```

```python
find_word(board, word)
```
