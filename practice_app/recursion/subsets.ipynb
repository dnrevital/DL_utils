```python
nums = [1, 2, 3, 4]
```

```python
def find_subsets(nums):
    if not nums:
        return []
    
    subsets = []
    
    for i in range(len(nums)):
        subset = nums.copy()
        subset.pop(i)
        found = find_subsets(subset)

        for item in found:
            if item not in subsets:
                subsets.append(item)
    
    subsets.append(nums)
    
    return subsets
```

```python
find_subsets(nums)
```
