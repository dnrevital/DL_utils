```python
matrix = [[9,9,4],
          [6,6,8],
          [2,1,1]]
```

```python
def find_path_len(matrix, i, j, path_lens):
    if path_lens[i][j] > 0:
        return path_lens[i][j]
    
    rows = len(matrix)
    cols = len(matrix[0])
    nbrs = []
    my_path_lens = []
    
    if i > 0:
        nbrs.append([i-1, j])
    if i < rows-1:
        nbrs.append([i+1, j])
    if j > 0:
        nbrs.append([i, j-1])
    if j < cols-1:
        nbrs.append([i, j+1])  
        
    for nbr in nbrs:
        if matrix[nbr[0]][nbr[1]] > matrix[i][j]:
            my_path_lens.append(find_path_len(matrix, nbr[0], nbr[1], path_lens))
     
    if not my_path_lens:
        path_lens[i][j] = 1
    else: 
        path_lens[i][j] = max(my_path_lens) + 1
        
    return path_lens[i][j]
```

```python
def longest_increasing_path(matrix):
    max_path_len = 0
    rows = len(matrix)
    cols = len(matrix[0])
    path_lens = [[0 for j in range(cols)] for i in range(rows)]
    
    for i in range(rows):
        for j in range(cols):
            if path_lens[i][j] == 0:
                path_len = find_path_len(matrix, i, j, path_lens) 
                if path_len > max_path_len:
                    max_path_len = path_len
                    
    return max_path_len
```

```python
matrix = [[9,9,4],
          [6,6,8],
          [2,1,1]]
```

```python
longest_increasing_path(matrix)
```
