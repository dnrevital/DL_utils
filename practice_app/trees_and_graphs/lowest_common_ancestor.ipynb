```python
from collections import defaultdict
```

```python
class Node:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None
```

```python
nodes = [3,5,1,6,2,0,8,None,None,7,4]
```

```python
tree = Node(3)
tree.left = Node(5)
tree.right = Node(1)
tree.left.left = Node(6)
tree.left.right = Node(2)
tree.right.left = Node(0)
tree.right.right = Node(8)
tree.left.right.left = Node(7)
tree.left.right.right = Node(4)
```

```python
def build_ancestors(root, ancestors):
    if root.left:
        ancestors[root.left.val] = ancestors[root.val] + [root.val]
        build_ancestors(root.left, ancestors)
    if root.right:
        ancestors[root.right.val] = ancestors[root.val] + [root.val]
        build_ancestors(root.right, ancestors)        
```

```python
def lca(root, p, q):
    ancestors = defaultdict(list)
    build_ancestors(root, ancestors)
    
    print(f'ancestors[{p}]: {ancestors[p]}')
    print(f'ancestors[{q}]: {ancestors[q]}')
    
    for i in range (len(ancestors[p])-1, -1, -1):
        if ancestors[p][i] in ancestors[q]:
            return ancestors[p][i]
```

```python
lca(tree, 7, 0)
```
