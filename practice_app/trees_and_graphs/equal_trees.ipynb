```python
class Node:
    def __init__(self, item):
        self.item = item
        self.left = None
        self.right = None 
```

```python
t1 = Node(1)
t1.left = Node(2)
t1.right = Node(3)

t2 = Node(1)
t2.left =  Node(3)
t2.right = Node(2)
```

```python
def is_equal(t1, t2):
    if (not t1) and (not t2):
        return True
    
    if t1 and t2 and t1.item == t2.item:
        return is_equal(t1.left, t2.left) and is_equal(t1.right, t2.right)
        
    return False
```

```python
is_equal(t1, t2)
```
