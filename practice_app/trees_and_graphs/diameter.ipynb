```python
class Node:
    def __init__(self, item):
        self.item = item
        self.left = None
        self.right = None 
```

```python
tree = Node(1)
tree.left = Node(2)
tree.left.left = Node(4)
tree.left.right = Node(5)
tree.right = Node(3)
```

```python
def max_dist(root):
    d1, d2 = 0, 0
    
    if root.left:
        d1 = max_dist(root.left)
    if root.right:
        d2 = max_dist(root.right)
    
    return max(d1, d2) + 1
```

```python
def diameter(root):
    d1, d2 = 0, 0
    
    if root.left:
        d1 = max_dist(root.left)
    if root.right:
        d2 = max_dist(root.right)
        
    return d1 + d2
```

```python
diameter(tree)
```
