```python
class Node:
    def __init__(self, item):
        self.item = item
        self.left = None
        self.right = None 
```

```python
t = Node(3)
t.left = Node(9)
t.right = Node(20)
t.right.left = Node(15)
t.right.right = Node(7)
```

```python
def max_depth(t, depth):
    depth += 1
    d1, d2 = 0, 0
    
    if not t.left and not t.right:
        return depth
    
    if t.left:
        d1 = max_depth(t.left, depth)
    if t.right:
        d2 = max_depth(t.right, depth)
    
    return max(d1, d2)
```

```python
max_depth(t, 0)
```
