```python
nums = [37, 44, 55, 61, 77, 78, 87, 89, 2, 5, 12, 21]
target = 77
```

```python
def rot_search(nums, target):
    pivot = 0
    for i in range(1, len(nums)):
        if nums[i] < nums[i-1]:
            pivot = i
            
    tot_len = len(nums)
    rot_left = pivot
    rot_right = pivot-1
    left = rot_left - pivot
    right = rot_right + pivot
    items = right - left + 1
    
    while right > left:
        cur = left + items//2
        rot_cur = (cur+pivot)%tot_len
        if nums[rot_cur] == target:
            return rot_cur
        if nums[rot_cur] > target:
            left = max(left-1, 0)
        elif nums[rot_cur] < target:
            right = min(right+1, tot_len-1)
        items = right - left + 1
        
    rot_left = (left+pivot)%tot_len
    rot_right = (right+pivot)%tot_len
    
    if nums[rot_left] == target:
        return rot_left
    if nums[rot_right] == target:
        return rot_right
    return -1
```

```python
rot_search(nums, target)
```
