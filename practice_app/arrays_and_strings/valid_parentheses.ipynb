```python
s1 = '([)]'
s2 = '{[]}'
```

```python
def validate(par):
    s = []
    left = '([{'
    right = ')]}'
    paired = {')': '(', ']': '[', '}': '{'}
    
    for c in par:
        if c in left:
            s.insert(0, c)
        elif c in right:
            if s[0] != paired[c]:
               return False
            s.pop(0)
            
    return True
```

```python
validate(s1)
```

```python
validate(s2)
```
