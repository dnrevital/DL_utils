```python
import numpy as np
```

```python
matrix = [[1,2,3],
          [4,5,6],
          [7,8,9]]
```

```python
def spiral(matrix):
    rows = len(matrix)
    cols = len(matrix[0])
    visited = [[False for j in range(cols)] for i in range(rows)]
    spiral_ = []
    rec(matrix, 0, 0, visited, spiral_)
    
    return spiral_

def rec(matrix, i, j, visited, spiral):
    visited[i][j] = True
    spiral.append(matrix[i][j])
    rows = len(matrix)
    cols = len(matrix[0])
    nexts = [[i, j+1], [i+1, j], [i, j-1], [i-1, j]]
    avail = [n[0]>=0 and n[0]<rows and n[1]>=0 and n[1]<cols for n in nexts]
    next_avails = [n for i, n in enumerate(nexts) if avail[i]]
    next_avail_not_visited = [n for n in next_avails if not visited[n[0]][n[1]]]
    if not next_avail_not_visited:
        return
    next_ = next_avail_not_visited[0]
    rec(matrix, next_[0], next_[1], visited, spiral)
```

```python
spiral(matrix)
```
