```python
nums = [30, 2, 40, 55, 7, 11, 15]
target = 9
```

```python
nums_inds = {n: i for i, n in enumerate(nums)}
```

```python
diffs = [target-n for n in nums]
```

```python
def two_sum(nums, target):
    nums_inds = {n: i for i, n in enumerate(nums)}
    diffs = [target-n for n in nums]
    
    for i, d in enumerate(diffs):
        j = nums_inds.get(d)
        if j:
            return [i, j]
        
    return -1
```

```python
two_sum(nums, target)
```
