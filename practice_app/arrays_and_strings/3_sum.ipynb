```python
nums = [-1,0,1,2,-1,-4]
```

```python
def three_sum(nums):
    two_sums = {}
    nnums = len(nums)
    
    for i in range(nnums):
        for j in range(i+1, nnums):
            two_sums[nums[i] + nums[j]] = [i, j]
            
    three_sums = []
    three_sets = []
            
    for i, n in enumerate(nums):
        inds = two_sums.get(-n)
        if inds:
            three_set = set({nums[inds[0]], nums[inds[1]], n})
            if i not in inds and three_set not in three_sets:
                three_sums.append([nums[inds[0]], nums[inds[1]], n])
                three_sets.append(three_set)
        
    return three_sums
```

```python
three_sum(nums)
```
