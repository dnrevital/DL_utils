```python
nums = [2, 5, 12, 21, 37, 44, 55, 61, 77, 78, 87, 89]
target = 77
```

```python
def bsearch(nums, target):
    left = 0
    right = len(nums) - 1
    items = right - left + 1
    
    while right > left:
        cur = left + items//2
        if nums[cur] == target:
            return cur
        if nums[cur]>target:
            right = max(cur-1, 0)
        elif nums[cur]<target:
            left = min(cur+1, len(nums)-1)
        items = right - left + 1
        
    if nums[left] == target:
        return left
    if nums[right] == target:
        return right
    return -1
```

```python
bsearch(nums, target)
```
