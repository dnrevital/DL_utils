```python
s = 'abc1def'
exp = 'abc.def'
```

```python
def is_matching(s, exp):
    if not s:
        return exp in ['', '*']
    
    if len(s) == 1:
        return len(exp) == 1 and (s[0] == exp[0] or exp[0] in ['.', '*'])
    
    if exp[0] not in ['.', '*']:
        return s[0] == exp[0] and is_matching(s[1:], exp[1:])
    
    if exp[0] == '.':
        return is_matching(s[1:], exp[1:])
    
    if exp[0] == '*':
        return len(exp) == 1 or is_matching(s, exp[1:]) or is_matching(s[1:], exp)
```

```python
is_matching(s, exp)
```
