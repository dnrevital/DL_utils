```python
prices = [7,1,5,3,6,4]
```

```python
def best_profit(prices):
    min_price = prices[0]
    max_profit = 0
    
    for price in prices[1:]:
        profit = price - min_price
        if profit > max_profit:
            max_profit = profit
        elif price < min_price:
            min_price = price
        
    return max_profit
```

```python
best_profit(prices)
```
