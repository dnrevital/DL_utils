```python
nums = [-2,1,-3,4,-1,2,1,-5,4]
#nums = [1, 1, 1]
```

```python
def max_subarray(nums):
    size = len(nums)
    sums = [[0 for j in range(size)] for i in range(size)]
    
    for i in range(size):
        sums[i][i] = nums[i]
    
    for i in range(size):
        for j in range(i+1, size):
            sums[i][j] = sums[i][j-1] + nums[j]
            
    print(f'sums: {sums}')
    
    return max([max(row) for row in sums])
```

```python
max_subarray(nums)
```
