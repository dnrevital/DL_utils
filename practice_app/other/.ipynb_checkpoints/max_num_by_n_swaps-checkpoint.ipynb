```python
from collections import defaultdict
```

```python
def max_by_swaps(num, swaps):
    digit_inds = defaultdict(list)
    digits = [int(digit) for digit in str(num)[:swaps+1]]
    leftover_digits = [int(digit) for digit in str(num)[swaps+1:]]
    
    for i, digit in enumerate(digits):
        digit_inds[digit].append(i)

    rem = swaps
    biggest_digits = []

    while rem:
        max_digit = max(digits)
        ind = digit_inds[max_digit][0]
        if ind <= rem:
            biggest_digits.append(max_digit)
            digits.pop(ind)
            digit_inds[max_digit].pop(0)
        rem -= ind
            
    res_digits = biggest_digits + digits + leftover_digits
    return int(''.join([str(i) for i in res_digits]))
```

```python
num = 12345
swaps = 2
```

```python
max_by_swaps(num, swaps)
```
