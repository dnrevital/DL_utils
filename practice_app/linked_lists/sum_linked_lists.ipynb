```python
class Linked:
    def __init__(self, n):
        self.n = n
        self.next = None
        
    def insert(self, n):
        self.next = Linked(n)
        
    def print(self):
        cur = self
        while cur:
            print(f'n: {cur.n} ==> ', end='')
            cur = cur.next            
        print()
        
    def full_num(self):
        n = self.n
        mul = 10
        cur = self
        
        while cur.next:
            n += cur.next.n * mul
            mul *= 10
            cur = cur.next
            
        return n
        
```

```python
l1 = Linked(3)
l1.insert(2)
l1.next.insert(1)

l2 = Linked(6)
l2.insert(5)
l2.next.insert(4)
```

```python
l1.print(), l2.print()
```

```python
l1.full_num(), l2.full_num()
```

```python
def build_list(n):
    s = str(n)
    digits = [c for c in s]
    cur = Linked(digits[-1])
    first = cur
    
    for digit in digits[-2::-1]:
        cur.next = Linked(digit)
        cur = cur.next  
        
    return first
```

```python
build_list(l1.full_num() + l2.full_num()).print()
```
