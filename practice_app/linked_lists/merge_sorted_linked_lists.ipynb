```python
class Linked:
    def __init__(self, n=None):
        self.n = n
        self.next = None
        
    def insert(self, n):
        self.next = Linked(n)
        
    def print(self):
        cur = self
        while cur:
            print(f'n: {cur.n} ==> ', end='')
            cur = cur.next            
        print()        
```

```python
l1 = Linked(1)
l1.insert(2)
l1.next.insert(4)

l2 = Linked(1)
l2.insert(3)
l2.next.insert(4)
```

```python
l1.print(), l2.print()
```

```python
def merged(l1, l2):
    cur = Linked()
    first = cur
    last1 = l1
    last2 = l2
    
    while last1 and last2:
        cur.n = min(last1.n, last2.n)
        if last1.n < last2.n:
            last1 = last1.next
        else:
            last2 = last2.next
        cur.next = Linked()
        cur = cur.next
        
    while last1:
        cur.n = last1.n
        last1 = last1.next
        cur.next = Linked()
        cur = cur.next

    while last2:
        cur.n = last1.n
        last2 = last2.next
        cur.next = Linked()
        cur = cur.next
              
    return first
```

```python
merged(l1, l2).print()
```
