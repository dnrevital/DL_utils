## Imports

```python
import os
import io
import boto3
import cv2
from matplotlib import pyplot as plt
```

## Settings

```python
imdir = '../obstacles_classification_datasets/test_3/train'
imname = '43_obstacles_classification_10_2022-11-09T13-34-51_1_0.03_shufersal_obs_3.0000_.jpg'
impath = os.path.join(imdir, imname)
```

## Utils

```python
def display_image(im, title):
    plt.imshow(im)
    plt.title(title)
    plt.show()    
```

## Main

```python
im = cv2.imread(os.path.join(imdir, imname))
im = cv2.cvtColor(im, cv2.COLOR_BGR2RGB) 

# im0: Horizontal Flip
im0 = cv2.flip(im, 0)

# Divide im to original components (ref, current, mask)
w = im.shape[1]
ref = im[:, :w//3]
current = im[:, w//3:(w*2)//3] 
mask = im[:, (w*2)//3:] 

# im1: Vertical Flip
ref = cv2.flip(ref, 1)
current = cv2.flip(current, 1)
im1 = cv2.hconcat([ref, current, mask])

# Display Images
display_image(im, 'im')
display_image(im0, 'im0')
display_image(im1, 'im1')
```
