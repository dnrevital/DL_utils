## Imports

```python
import cv2
import os
import numpy as np
from matplotlib import pyplot as plt
```

# Settings

```python
folder = '/home/drevital/obstacles_classification_datasets/gg_2/obstacle_pairs_1'
imname = '9_746_.94_2021-08-15--16-35-24_.jpg'
```

## Load Image

```python
im = cv2.imread(os.path.join(folder,imname))
```

```python
im.shape
```

```python
im[300][:20]
```

## Split Images

```python
w = im.shape[1]
im1 = im[:,:w//2]
im2 = im[:,w//2:]
```

## Display Splitted Images

```python
plt.imshow(cv2.cvtColor(im1, cv2.COLOR_BGR2RGB))
plt.title('left half = source')
plt.show()

plt.imshow(cv2.cvtColor(im2, cv2.COLOR_BGR2RGB))
plt.title('right half = current')
plt.show()
```
