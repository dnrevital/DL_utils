## Imports

```python
import os
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

## Settings

```python
imdir = './lowest'
imname = 'v2_frame_000130.png'

# obb verices
x1 = 300
x2 = 400
x3 = 400
x4 = 300
y1 = 350
y2 = 350
y3 = 420
y4 = 420
```

## Read the image, draw the bbox

```python
# Load the image
img = cv2.imread(os.path.join(imdir, imname))

# Define the corner coordinates of the bounding box
pts = [[x1, y1], [x2, y2], [x3, y3], [x4, y4]]

# Convert the coordinates to a numpy array
pts = np.array(pts, dtype=np.int32)

# Draw the bounding box on the image
_ = cv2.polylines(img, [pts], True, (0, 255, 0), 2)

# Print the object's name inside the bounding box
object_name = 'forklift'

# Get the center of the bounding box
x = int((x1 + x2 + x3 + x4) / 4)
y = int((y1 + y2 + y3 + y4) / 4)

# Draw the object name on the image
_ = cv2.putText(img, object_name, (x, y), cv2.FONT_HERSHEY_SIMPLEX, 1.25, (0, 255, 0), 2)
```

## Display image with bbox

```python
plt.imshow(img)
plt.title('image with bbox')
plt.show()
```
