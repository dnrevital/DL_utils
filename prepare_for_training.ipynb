## Imports

```python
import cv2
import os
import numpy as np
from matplotlib import pyplot as plt
```

## Load Images Utility

```python
def load_images_from_folder(folder):
    imnames = []
    images = []
    for imname in os.listdir(folder):
        imrgb = cv2.imread(os.path.join(folder,imname))
        im = cv2.cvtColor(imrgb, cv2.COLOR_BGR2GRAY)        
        if im is not None:
            imnames.append(imname)
            images.append(im)
    return imnames, images
```

## Load & Process Loop

```python
obs_folder='/home/drevital/obstacle_detection_dataset/obstacle'
obs_imnames, obs_images = load_images_from_folder(obs_folder)
no_obs_folder='/home/drevital/obstacle_detection_dataset/not_obstacle'
no_obs_imnames, no_obs_images = load_images_from_folder(no_obs_folder)
```

## Generate Diff Obstacle Images and Store Them

```python
diffs_folder = '/home/drevital/obstacle_detection_dataset/obstacle_detection_diffs/obstacle/'

for i, imname in enumerate(obs_imnames):
    im = obs_images[i]
    w = im.shape[1]
    im1 = im[:,:w//2]
    im2 = im[:,w//2:]
    plain = cv2.subtract(im2, im1)
    cv2.imwrite(diffs_folder + imname, plain)
```

## Generate Diff Non-Obstacle Images and Store Them

```python
diffs_folder = '/home/drevital/obstacle_detection_dataset/obstacle_detection_diffs/no_obstacle/'

for i, imname in enumerate(no_obs_imnames):
    im = no_obs_images[i]
    w = im.shape[1]
    im1 = im[:,:w//2]
    im2 = im[:,w//2:]
    plain = cv2.subtract(im2, im1)
    cv2.imwrite(diffs_folder + imname, plain)
```
