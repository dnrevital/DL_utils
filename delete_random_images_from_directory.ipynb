## Imports

```python
import os
import numpy as np
from numpy.random import default_rng
```

## Settings

```python
from_dir = '/home/drevital/obstacles_classification_datasets/3_class/pairs/obstacle_1'
num_deleted = 409
```

## Main

```python
fnames = os.listdir(from_dir)
num_files = len(fnames)
rng = default_rng()
rands = rng.choice(num_files, size=num_deleted, replace=False)
```

```python
for i in rands:
    fname = os.path.join(from_dir, fnames[i])
    try:
        os.remove(fname)
    except Exception as e:
        print(e)
```
